## 缺失值处理

### 缺失值统计

```python
nans = pd.concat([data.isnull().sum(), data.isnull().sum() / data.shape[0]],
                 axis=1, keys=['Data', 'Percentage'])
nans[nans.sum(axis=1) > 0].sort_index(by=['Data'],ascending=False)
```

由此可见，许多值确实较多

一般而言，对于此处的数字类型（int64,float32）等数据直接填零比较合适（例如，student_feature的1.0代表是学生，而NaN应该就不是，此处可以用0进行表示）

```python
numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerics = []
for i in data.columns:
    if data[i].dtype in numeric_dtypes:
        numerics.append(i)
data.update(data[numerics].fillna(0))
```

```python
nans = pd.concat([data.isnull().sum(), data.isnull().sum() / data.shape[0]],
                 axis=1, keys=['Data', 'Percentage'])
nans[nans.sum(axis=1) > 0].sort_index(by=['Data'],ascending=False)
```

大部分数字类型的缺失值填补完毕后就只剩下时间和地理位置没填

```python
# 时间没填应该是初次借款，统一填为None
data.update(data['latest_query_time'].fillna('None'))
data.update(data['loans_latest_time'].fillna('None'))
```

```python
# 地理位置编码为四种情况，一线，二线，三线，境外，分别对应1,2,3,4
# 缺失值填为三线城市，因为小城市的人更不愿意承认自己来自小城市
for i in range(4754):
    if data['reg_preference_for_trad'][i]=='一线城市'.encode('gbk').decode('gbk'):
        data['reg_preference_for_trad'][i]=1
    elif data['reg_preference_for_trad'][i]=='二线城市'.encode('gbk').decode('gbk'):
        data['reg_preference_for_trad'][i]=2
    elif data['reg_preference_for_trad'][i]=='三线城市'.encode('gbk').decode('gbk'):
        data['reg_preference_for_trad'][i]=3
    elif data['reg_preference_for_trad'][i]=='境外'.encode('gbk').decode('gbk'):
        data['reg_preference_for_trad'][i]=4
    else:
        data['reg_preference_for_trad'][i]=3
data.sample(10)
```
